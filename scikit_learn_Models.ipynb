```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression
from skmultilearn.problem_transform import BinaryRelevance
from sklearn.naive_bayes import GaussianNB
from skmultilearn.problem_transform import ClassifierChain
from sklearn.linear_model import SGDClassifier
from sklearn.svm import LinearSVC
from sklearn import tree
from sklearn.naive_bayes import MultinomialNB
from skmultilearn.problem_transform import LabelPowerset
from scipy.sparse import csr_matrix, lil_matrix
from skmultilearn.adapt import MLkNN
from sklearn.metrics import accuracy_score, f1_score, hamming_loss, precision_score, recall_score, classification_report
import warnings
warnings.filterwarnings('ignore')
```

```python
#data
data = pd.read_csv('/content/gdrive/My Drive/data.csv',index_col= None, delimiter=";")
data.tail(10).style.background_gradient(cmap='Wistia')
```

```python
#splitting & TF-IDF vectorizing
train, test = train_test_split(data, random_state=2020, test_size=0.20, shuffle = True)
vectorizer = TfidfVectorizer(strip_accents='unicode', analyzer='word', ngram_range=(1,2), norm='l2')
vectorizer.fit(train)
vectorizer.fit(test)
x_train = vectorizer.transform(train.Keyword)
y_train = train.drop(labels = ['Keyword'], axis=1)
x_test = vectorizer.transform(test.Keyword)
y_test = test.drop(labels = ['Keyword'], axis=1)
del data
```

```python
#train & evaluvation metrics
def metrics(mdl):
    model.fit(x_train, y_train)
    predictions = model.predict(x_test)
    print("\u0332".join(mdl))
    print("\nAccuracy = ",accuracy_score(y_test,predictions))
    print("Hamming loss ",hamming_loss(y_test,predictions))
    precision = precision_score(y_test, predictions, average='micro')
    recall = recall_score(y_test, predictions, average='micro')
    f1 = f1_score(y_test, predictions, average='micro')
    print("\nMicro-average quality numbers")
    print("Precision: {:.4f}, Recall: {:.4f}, F1-measure: {:.4f}".format(precision, recall, f1))
    precision = precision_score(y_test, predictions, average='macro')
    recall = recall_score(y_test, predictions, average='macro')
    f1 = f1_score(y_test, predictions, average='macro')
    print("\nMacro-average quality numbers")
    print("Precision: {:.4f}, Recall: {:.4f}, F1-measure: {:.4f}".format(precision, recall, f1))
    print("\nClassification report\n",classification_report(y_test,predictions))
```

```python
#OneVsRest Logistic Regression
model = OneVsRestClassifier(LogisticRegression(), n_jobs = -1)
metrics('OneVsRest - Logistic Regression')
```

```python
#OneVsRest Multinomial Naive Bayes classifier
model = OneVsRestClassifier(MultinomialNB(), n_jobs = -1)
metrics('OneVsRest Multinomial Naive Bayes classifier')
```

```python
#OneVsRest SGD Classifier
model = OneVsRestClassifier(SGDClassifier(loss='log', alpha=0.00001, penalty='l2'))
metrics('OneVsRest SGD Classifier')
```

```python
#OneVsRest Support Vector Classifier
model = OneVsRestClassifier(LinearSVC())
metrics('OneVsRest Support Vector Classifier')
```

```python
#Binary Relevance
model = BinaryRelevance(GaussianNB())
metrics('Binary Relevance')
```

```python
#Classifier Chains
model = ClassifierChain(LogisticRegression())
metrics('Classifier Chains')
```

```python
#Decision Tree Classifier
model = tree.DecisionTreeClassifier()
metrics('Decision Tree Classifier')
```

```python
#Label Powerset
model = LabelPowerset(LogisticRegression())
metrics('Label Powerset')
```
